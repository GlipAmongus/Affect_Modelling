```python
import numpy as np
import pandas as pd

from sklearn.preprocessing import StandardScaler
```

```python
# Replace with your actual data loading
data = pd.read_csv("./data./recola_data.csv")

print(data.columns)
print(data.head())
```

```python
# Separate features and targets
exclude_cols = ["Participant", "median_arousal", "median_valence"]

feature_columns = [col for col in data.columns if col not in exclude_cols]

X = data[feature_columns].values

# Targets
y_arousal = data["median_arousal"].values
y_valence = data["median_valence"].values

# Groups
participant_ids = data["Participant"].values

# Sanity check
print("X shape:", X.shape)
print("Num features:", len(feature_columns))
print("NAN:", np.isnan(X).sum())
```

```python
# Standardise features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

np.save("features/raw_features.npy", X_scaled)    
```

```python
np.save("features/y_reg_arousal.npy", y_arousal)
np.save("features/y_reg_valence.npy", y_valence)
```

```python
arousal_median = np.median(y_arousal)

y_clf_arousal = (y_arousal > arousal_median).astype(int)
```

```python
valence_median = np.median(y_valence)

y_clf_valence = (y_valence > valence_median).astype(int)
```

```python
np.save("features/y_clf_arousal.npy", y_clf_arousal)
np.save("features/y_clf_valence.npy", y_clf_valence)
```

```python
np.save("features/participant_ids.npy", participant_ids)
```

```python
print("Raw features shape:", X_scaled.shape)
print("Arousal classes:", np.bincount(y_clf_arousal))
print("Valence classes:", np.bincount(y_clf_valence))
```

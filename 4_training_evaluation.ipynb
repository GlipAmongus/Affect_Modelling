```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.neural_network import MLPRegressor, MLPClassifier

from sklearn.metrics import mean_squared_error, r2_score
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix
```

```python
# Example placeholder
# Replace with your actual dataset loading

data = pd.read_csv("dataset.csv")

# Example split
X = data.drop(columns=["regression_target", "classification_target"])
y_reg = data["regression_target"]
y_clf = data["classification_target"]
```

```python
X_train, X_test, y_reg_train, y_reg_test = train_test_split(
    X, y_reg, test_size=0.2, random_state=42
)

_, _, y_clf_train, y_clf_test = train_test_split(
    X, y_clf, test_size=0.2, random_state=42
)
```

```python
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
lin_reg = LinearRegression()
lin_reg.fit(X_train_scaled, y_reg_train)

y_pred = lin_reg.predict(X_test_scaled)

print("Linear Regression (Raw Features)")
print("MSE:", mean_squared_error(y_reg_test, y_pred))
print("R2:", r2_score(y_reg_test, y_pred))
```

```python
nn_reg = MLPRegressor(
    hidden_layer_sizes=(100,50),
    max_iter=500,
    random_state=42
)

nn_reg.fit(X_train_scaled, y_reg_train)

y_pred = nn_reg.predict(X_test_scaled)

print("Neural Network Regressor (Raw Features)")
print("MSE:", mean_squared_error(y_reg_test, y_pred))
print("R2:", r2_score(y_reg_test, y_pred))
```

```python
log_reg = LogisticRegression(max_iter=1000)

log_reg.fit(X_train_scaled, y_clf_train)

y_pred = log_reg.predict(X_test_scaled)

print("Logistic Regression (Raw Features)")
print("Accuracy:", accuracy_score(y_clf_test, y_pred))
print("F1:", f1_score(y_clf_test, y_pred, average='weighted'))
print("Confusion Matrix:\n", confusion_matrix(y_clf_test, y_pred))
```

```python
nn_clf = MLPClassifier(
    hidden_layer_sizes=(100,50),
    max_iter=500,
    random_state=42
)

nn_clf.fit(X_train_scaled, y_clf_train)

y_pred = nn_clf.predict(X_test_scaled)

print("Neural Network Classifier (Raw Features)")
print("Accuracy:", accuracy_score(y_clf_test, y_pred))
print("F1:", f1_score(y_clf_test, y_pred, average='weighted'))
print("Confusion Matrix:\n", confusion_matrix(y_clf_test, y_pred))
```
